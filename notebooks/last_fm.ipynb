```python
#/2.0/?method=chart.gettopartists&api_key=YOUR_API_KEY&format=json
api_chart = f"http://ws.audioscrobbler.com/2.0/?method=chart.gettopartists&api_key={apikey}&format=json"
top_art = requests.get(api_chart).json()
```
