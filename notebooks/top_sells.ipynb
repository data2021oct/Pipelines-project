```python
url_insider = "https://www.businessinsider.com/50-best-selling-albums-all-time-2016-9"
html_insider = requests.get(url_insider)
soup_insider = BeautifulSoup(html_insider.content,"html.parser")
```

```python
tags_insider = soup_insider.find_all("div", {"class": "slide-layout"})
```

```python
tags_insider
```

```python

"""
'\n\n50. Led Zeppelin —\xa0"Led Zeppelin II"\n\n\n\n\n\n\n\n\n\n\n                                                  Atlantic\n                                                \n\n\nCertified units: 12\xa0millionBuy it here >>\n\n\n'
"""
tags_insider[0].find_all("p")[0].getText() #'Certified units: 12\xa0million'
```

```python
insider = tags_insider[0].find("h2",{ "class":"slide-title-text"})
```

```python
insider.text
```

```python
tags_insider[0].find("h2").getText()
tags_insider[0].find_all("p")[0].getText()
```

```python
tags_insider[0]
```

```python
insider_list = []
for t in tags_insider:
    dicc_t = {"album": t.find("h2").getText(),
             "certified units" : t.find("p").getText()}
    insider_list.append(dicc_t)
```

```python
insider = pd.DataFrame(insider_list)
```

```python
insider.tail(15)
```

```python
insider["Rank"] = insider["album"].str.extract(r"(^\d+)")
```

```python
insider["Millions"] = insider["certified units"].str.extract(r"(\d+)")
```

```python
insider["album"] = insider["album"].str.replace(" – "," — ",regex = True)
```

```python
insider["Artist"] = insider["album"].str.extract(r"(^\d+.\s(.+?)—)")[1]
```

```python
insider["Title"] = insider["album"].str.extract(r"(—\s(.+?)$)")[1]
```

```python
insider["Title"] = insider["Title"].str.replace('"',"")
```

```python
insider[["Millions","Rank"]] = insider[["Millions","Rank"]].astype("int64")
```

```python
top_sells = insider[["Rank","Artist","Title","Millions"]]
```

```python
top_sells.head(5)
```

```python
apikey = os.getenv("apikey")
```

```python
api_urls = []
for s in range(len(top_sells)):

    apikey = os.getenv("apikey")
    artist = top_sells.loc[s,'Artist']
    album = top_sells.loc[s,'Title']
    api_urls.append(f"http://ws.audioscrobbler.com/2.0/?method=album.getinfo&api_key={apikey}&artist={artist}&album={album}&format=json")
    
```

```python
request_dic = [] 
i = 0
for a in api_urls:
    #request_dic.append(requests.get(a).json())
    res = requests.get(a).json()
    request_dic.append(res["album"])
    if i%10==0 or i== 49:
        print(f"{i} done")
    i+=1
```

```python
req_top_sales = pd.DataFrame(request_dic)
req_top_sales.sample(3)
```

```python
req_top_sales = req_top_sales[["artist","name","playcount","listeners"]]
```

```python
req_top_sales.sample(3)
```

```python
top_sells_rich = top_sells.merge(req_top_sales,left_index=True, right_index=True)
top_sells_rich.sample(10)
```

```python
top_sells_rich = top_sells_rich[["Rank","Artist","Title","playcount","listeners"]]
```
